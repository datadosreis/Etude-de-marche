### Scaling

Comme nous l'avons vu prcdemment lors de l'analyse de notre jeu de donnes, nous avons des distributions qui sont trs diffrentes selon les variables. Il est donc important de normaliser les valeurs des diffrentes colonnes afin de les rendre plus comparables et d'en faciliter le traitement.

Ici, nous allons utiliser l'objet *StandardScaler* qui est fourni par la bibliothque `scikit-learn`. Il permet de standardiser les donnes en calculant la moyenne et l'cart-type de chaque variable d'un jeu de donnes. Il utilise ensuite ces valeurs pour centrer et rduire chaque variable en fonction de ces statistiques. Cela permet de mettre toutes les variables sur la mme chelle, avec une moyenne de 0 et une variance de 1.

```python
#On ralise une copie df avant de scaler les donnes
data_scaled_secto = df_sectoriel.copy()

#On enregistre les colonnes dans une variable
features_secto = data_scaled_secto.columns

# initialisation de l'objet StandardScaler
sc = StandardScaler()

# fit_transform sur les colonnes slectionnes
data_scaled_secto[features_secto] = sc.fit_transform(data_scaled_secto[features_secto])
```

```python
#Vrification des rsultats du scaling
idx = ["mean", "std", "min","50%","max"]
pd.DataFrame(data_scaled_secto).describe().round(2).loc[idx, :]
```

#### Mthode 2: lalgorithme k-means

```python
#Copie du df sectoriel avec donnes scales
X_secto = data_scaled_secto.copy()
```

L'algorithme du k-means impose de dfinir  l'avance le nombre de clusters souhait.
Afin de dterminer le nombre optimal de clusters, il existe deux mthodes que nous allons utiliser ci-dessous:

- La mthode du coude
- Le silhouette score

##### Mthode du coude

```python
#Cration d'une liste vide pour stocker les inerties
inertie = []

#On dfinit dans une liste le nombre de clusters que l'on souhaite tester
k_list = range(1, 10)

#Pour chaque valeur de k, on entrane un k-means spcifique et l'on stocke son inertie dans la liste cre au pralable

for k in k_list :
  kmeans_secto = KMeans(n_clusters=k)
  kmeans_secto.fit(X_secto) 
  inertie.append(kmeans_secto.inertia_)

#Cration du graphique pour afficher le rsultat
fig, ax = plt.subplots(1,1,figsize=(10,3))

ax.set(title="Mthode du coude", ylabel = "Inertie",xlabel = "Nb cluster")
ax = plt.plot(k_list, inertie, marker = "o")
```

Le nombre de clusters optimal est gnralement considr comme celui o l'on observe un "coude" dans la courbe. Cela signifie qu'il y a un point o l'inertie ralentit considrablement, ce qui indique qu'ajouter de nouveaux clusters n'apporte pas beaucoup de bnfice supplmentaire en termes de sparation des donnes.

##### Silhouette score

```python
#Cration d'une liste vide pour stocker les cofs
silhouette = []

#On dfinit dans une liste le nombre de clusters que l'on souhaite tester
i_list = range(2, 10)

#Pour chaque valeur de k, on entrane un k-means spcifique et l'on stocke son cof dans la liste cre au pralable

for i in i_list :
  kmeans_secto = KMeans(n_clusters=i)
  kmeans_secto.fit(X_secto) 
  silhouette_score = metrics.silhouette_score(X_secto, kmeans_secto.labels_)
  silhouette.append(silhouette_score)

#Cration du graphique pour afficher le rsultat
fig, ax = plt.subplots(1,1,figsize=(10,3))

ax.set(title="Silhouette Score", ylabel = "Score",xlabel = "Nombre de clusters")
ax = plt.plot(i_list, silhouette, marker = "o")
```

Le nombre optimal de clusters est de 4.

### Projection des donnes via ACP

```python
#On enregistre les clusters dans un df
X_cluster_secto = X_secto["Cluster"].to_frame()
#On supprime la colonne "Cluster" en prvision de l'ACP
X_secto.drop(columns="Cluster",inplace=True)
```

On peut maintenant raliser notre ACP. On instancie l'objet PCA de la bibliothque `scikit-learn` et on entraine notre ACP.

```python
#Choix du nombre de composantes
n_components_secto = 4

pca_secto = PCA(n_components=n_components_secto)
pca_secto.fit(data_scaled_secto)
```

#### Eboulis des valeurs propres

On peut calculer notre variance cumule et effectuer une visualisation:

```python
#Calcul de la variance cumule
scree_secto = (pca_secto.explained_variance_ratio_*100).round(2)
scree_cum_secto = scree_secto.cumsum().round()
x_list_secto = range(1, n_components_secto+1)

#Reprsentation graphique
plt.figure(figsize=(10,3))
plt.bar(x_list_secto, scree_secto)
plt.plot(x_list_secto, scree_cum_secto,c="red",marker='o')
plt.xlabel("Rang de l'axe d'inertie")
plt.ylabel("Pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)
```

On voit ici que plus de 80% de la variance est comprise dans les 2 premires composantes.

#### Cercle des corrlations

Intressons nous dans un premier temps aux diffrentes composantes.

```python
#Calcul des cofficients des variables pour chaque composante
pcs_secto = pca_secto.components_
#On met ces valeurs dans un df
pcs_secto = pd.DataFrame(pcs_secto)
#On rajoute les colonnes
pcs_secto.columns = X_secto.columns
pcs_secto.index = [f"PC{i}" for i in x_list_secto]
#On transpose pour une meilleure visibilit
pcs_secto = pcs_secto.T
pcs_secto
```

```python
# Heatmap
sns.heatmap(pcs_secto, cmap="YlGnBu")

plt.xlabel("Composantes")
plt.ylabel("Variables")

plt.show()
```

```python
#On enregistre les colonnes dans une variable
features_acp_secto = pcs_secto.columns
```

Dfinissons nos axes x et y. Nous allons ici utiliser les 2 premires composantes.

```python
x, y = 0,1
```

On peut maintenant crer le cercle des corrlations.

```python
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(0, pca_secto.components_.shape[1]):
    ax.arrow(0,
             0,  # Dbut de la flche  l'origine
             pca_secto.components_[x, i],  #0 pour PC1
             pca_secto.components_[y, i],  #1 pour PC2
             head_width=0.05,
             head_length=0.05, 
             width=0.02,              )

    plt.text(pca_secto.components_[x, i] + 0.05,
             pca_secto.components_[y, i] + 0.05,
             X_secto.columns[i])
    
# affichage des lignes horizontales et verticales
plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')


# nom des axes, avec le pourcentage d'inertie expliqu
plt.xlabel('PC{} ({}%)'.format(x+1, round(100*pca_secto.explained_variance_ratio_[x],1)))
plt.ylabel('PC{} ({}%)'.format(y+1, round(100*pca_secto.explained_variance_ratio_[y],1)))

# titre
plt.title("Cercle des corrlations (PC{} et PC{})".format(x+1, y+1), fontsize=16, fontweight="bold")

#ajout du cercle
plt.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), np.sin(np.linspace(0, 2 * np.pi, 100)), linewidth=2)

plt.show()
```

Nous avons ici la premire composante PC1 qui est une variable synthtique des variables nourriture, production et importations.
La deuxime composante est elle reprsentative de la variable "%Evolution de la nourriture".

### Analyse en Composantes Principales

```python
#On enregistre les clusters dans un df
X_cluster = X["Cluster"].to_frame()
#On supprime la colonne "Cluster" en prvision de l'ACP
X.drop(columns="Cluster",inplace=True)
```

On peut maintenant faire notre ACP.

```python
#Choix du nombre de composantes
n_components = 10

pca = PCA(n_components=n_components)
pca.fit(X)
```

#### Eboulis des valeurs propres

```python
pca.explained_variance_ratio_
```

On peut calculer notre variance cumule et effectuer une visualisation:

```python
#Calcul de la variance cumule
scree = (pca.explained_variance_ratio_*100).round(2)
scree_cum = scree.cumsum().round()
x_list = range(1, n_components+1)

#Reprsentation graphique
plt.figure(figsize=(10,3))
plt.bar(x_list, scree)
plt.plot(x_list, scree_cum,c="red",marker='o')
plt.xlabel("Rang de l'axe d'inertie")
plt.ylabel("Pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)
```

On voit ici que les deux premires composantes rassemblent un peu plus de 50% de la variance.

#### Cercle des corrlations

Intressons nous dans un premier temps aux diffrentes composantes.

```python
#Calcul des cofficients des variables pour chaque composante
pcs = pca.components_
#On met ces valeurs dans un df
pcs = pd.DataFrame(pcs)
#On rajoute les colonnes
pcs.columns = X.columns
pcs.index = [f"PC{i}" for i in x_list]
#On transpose pour une meilleure visibilit
pcs = pcs.T
pcs
```

```python
#Heatmap
sns.heatmap(pcs, cmap="YlGnBu")

plt.xlabel("Composantes")
plt.ylabel("Variables")

plt.show()
```

```python
#On enregistre les colonnes dans une variable
features_acp = pcs.columns
```

Dfinissons nos axes x et y. 
Nous allons utiliser les 2 premires composantes.

```python
x, y = 0,1
```

Nous pouvons maintenant crer le cercle des corrlations.

```python
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Dbut de la flche  l'origine
             pca.components_[x, i],  #0 pour PC1
             pca.components_[y, i],  #1 pour PC2
             head_width=0.07,
             head_length=0.07, 
             width=0.02,              )

    plt.text(pca.components_[x, i] + 0.05,
             pca.components_[y, i] + 0.05,
             X.columns[i])
    
# affichage des lignes horizontales et verticales
plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')


# nom des axes, avec le pourcentage d'inertie expliqu
plt.xlabel('PC{} ({}%)'.format(x+1, round(100*pca.explained_variance_ratio_[x],1)))
plt.ylabel('PC{} ({}%)'.format(y+1, round(100*pca.explained_variance_ratio_[y],1)))

# titre
plt.title("Cercle des corrlations (PC{} et PC{})".format(x+1, y+1), fontsize=16, fontweight="bold")

#ajout du cercle
plt.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), np.sin(np.linspace(0, 2 * np.pi, 100)), linewidth=2)

plt.show()
```

Regardons maintenant les composantes 3 et 4:

```python
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Dbut de la flche  l'origine
             pca.components_[2, i],  #0 pour PC1
             pca.components_[3, i],  #1 pour PC2
             head_width=0.07,
             head_length=0.07, 
             width=0.02,              )

    plt.text(pca.components_[2, i] + 0.05,
             pca.components_[3, i] + 0.05,
             X.columns[i])
    
# affichage des lignes horizontales et verticales
plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')


# nom des axes, avec le pourcentage d'inertie expliqu
plt.xlabel('PC{} ({}%)'.format(2+1, round(100*pca.explained_variance_ratio_[2],1)))
plt.ylabel('PC{} ({}%)'.format(3+1, round(100*pca.explained_variance_ratio_[3],1)))

# titre
plt.title("Cercle des corrlations (PC{} et PC{})".format(2+1, 3+1), fontsize=16, fontweight="bold")

#ajout du cercle
plt.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), np.sin(np.linspace(0, 2 * np.pi, 100)), linewidth=2)

plt.show()
```

Ici l'interprtation des diffrentes composantes est rendu difficile par le fait que nous ayons des variables trs diffrentes avec notamment des donnes quantitatives (production, importations, nourriture, PIB par habitant), des donnes mesures en % (taux d'inflation, taux de chomage), un indice et des volutions en % (population, consommation de volailles, PIB).
