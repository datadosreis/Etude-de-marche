## Première itération avec les données sectorielles uniquement

Dans un premier temps, nous souhaitons faire une première itération avec uniquement les données sectorielles. L'objectif est ici de simplifier le traitement en conservant uniquement les pays présentant un réel potentiel marché et en supprimant les autres.

### Préparation des données sectorielles

```python
#Création d'un df avec uniquement les données sectorielles
df_sectoriel = df_final[['Nourriture (2019)',
                     '%Evolution nourriture',
                     'Production',
                     'Importations']]
df_sectoriel.head()
```

### Scaling

Comme nous l'avons vu précédemment lors de l'analyse de notre jeu de données, nous avons des distributions qui sont très différentes selon les variables. Il est donc important de normaliser les valeurs des différentes colonnes afin de les rendre plus comparables et d'en faciliter le traitement.

Ici, nous allons utiliser l'objet *StandardScaler* qui est fourni par la bibliothèque `scikit-learn`. Il permet de standardiser les données en calculant la moyenne et l'écart-type de chaque variable d'un jeu de données. Il utilise ensuite ces valeurs pour centrer et réduire chaque variable en fonction de ces statistiques. Cela permet de mettre toutes les variables sur la même échelle, avec une moyenne de 0 et une variance de 1.

```python
#On réalise une copie df avant de scaler les données
data_scaled_secto = df_sectoriel.copy()

#On enregistre les colonnes dans une variable
features_secto = data_scaled_secto.columns

# initialisation de l'objet StandardScaler
sc = StandardScaler()

# fit_transform sur les colonnes sélectionnées
data_scaled_secto[features_secto] = sc.fit_transform(data_scaled_secto[features_secto])
```

```python
#Vérification des résultats du scaling
idx = ["mean", "std", "min","50%","max"]
pd.DataFrame(data_scaled_secto).describe().round(2).loc[idx, :]
```

```python
#Distribution des données après le scaling
plt.figure(figsize=(20,10))

for i, col in enumerate(data_scaled_secto):
    plt.subplot(2, 2, i+1)
    sns.histplot(data_scaled_secto[col],ec="black", kde=True)
    plt.title(col,fontsize=12)

plt.show()
```

#### Méthode 2: l’algorithme k-means

```python
#Copie du df sectoriel avec données scalées
X_secto = data_scaled_secto.copy()
```

L'algorithme du k-means impose de définir à l'avance le nombre de clusters souhaité.
Afin de déterminer le nombre optimal de clusters, il existe deux méthodes que nous allons utiliser ci-dessous:

- La méthode du coude
- Le silhouette score

##### Méthode du coude

```python
#Création d'une liste vide pour stocker les inerties
inertie = []

#On définit dans une liste le nombre de clusters que l'on souhaite tester
k_list = range(1, 10)

#Pour chaque valeur de k, on entraîne un k-means spécifique et l'on stocke son inertie dans la liste créée au préalable

for k in k_list :
  kmeans_secto = KMeans(n_clusters=k)
  kmeans_secto.fit(X_secto) 
  inertie.append(kmeans_secto.inertia_)

#Création du graphique pour afficher le résultat
fig, ax = plt.subplots(1,1,figsize=(10,3))

ax.set(title="Méthode du coude", ylabel = "Inertie",xlabel = "Nb cluster")
ax = plt.plot(k_list, inertie, marker = "o")
```

Le nombre de clusters optimal est généralement considéré comme celui où l'on observe un "coude" dans la courbe. Cela signifie qu'il y a un point où l'inertie ralentit considérablement, ce qui indique qu'ajouter de nouveaux clusters n'apporte pas beaucoup de bénéfice supplémentaire en termes de séparation des données.

##### Silhouette score

```python
#Création d'une liste vide pour stocker les coéfs
silhouette = []

#On définit dans une liste le nombre de clusters que l'on souhaite tester
i_list = range(2, 10)

#Pour chaque valeur de k, on entraîne un k-means spécifique et l'on stocke son coéf dans la liste créée au préalable

for i in i_list :
  kmeans_secto = KMeans(n_clusters=i)
  kmeans_secto.fit(X_secto) 
  silhouette_score = metrics.silhouette_score(X_secto, kmeans_secto.labels_)
  silhouette.append(silhouette_score)

#Création du graphique pour afficher le résultat
fig, ax = plt.subplots(1,1,figsize=(10,3))

ax.set(title="Silhouette Score", ylabel = "Score",xlabel = "Nombre de clusters")
ax = plt.plot(i_list, silhouette, marker = "o")
```

Le nombre optimal de clusters est de 4.

### Projection des données via ACP

```python
#On enregistre les clusters dans un df
X_cluster_secto = X_secto["Cluster"].to_frame()
#On supprime la colonne "Cluster" en prévision de l'ACP
X_secto.drop(columns="Cluster",inplace=True)
```

On peut maintenant réaliser notre ACP. On instancie l'objet PCA de la bibliothèque `scikit-learn` et on entraine notre ACP.

```python
#Choix du nombre de composantes
n_components_secto = 4

pca_secto = PCA(n_components=n_components_secto)
pca_secto.fit(data_scaled_secto)
```

#### Eboulis des valeurs propres

On peut calculer notre variance cumulée et effectuer une visualisation:

```python
#Calcul de la variance cumulée
scree_secto = (pca_secto.explained_variance_ratio_*100).round(2)
scree_cum_secto = scree_secto.cumsum().round()
x_list_secto = range(1, n_components_secto+1)

#Représentation graphique
plt.figure(figsize=(10,3))
plt.bar(x_list_secto, scree_secto)
plt.plot(x_list_secto, scree_cum_secto,c="red",marker='o')
plt.xlabel("Rang de l'axe d'inertie")
plt.ylabel("Pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)
```

On voit ici que plus de 80% de la variance est comprise dans les 2 premières composantes.

#### Cercle des corrélations

Intéressons nous dans un premier temps aux différentes composantes.

```python
#Calcul des coéfficients des variables pour chaque composante
pcs_secto = pca_secto.components_
#On met ces valeurs dans un df
pcs_secto = pd.DataFrame(pcs_secto)
#On rajoute les colonnes
pcs_secto.columns = X_secto.columns
pcs_secto.index = [f"PC{i}" for i in x_list_secto]
#On transpose pour une meilleure visibilité
pcs_secto = pcs_secto.T
pcs_secto
```

```python
# Heatmap
sns.heatmap(pcs_secto, cmap="YlGnBu")

plt.xlabel("Composantes")
plt.ylabel("Variables")

plt.show()
```

```python
#On enregistre les colonnes dans une variable
features_acp_secto = pcs_secto.columns
```

Définissons nos axes x et y. Nous allons ici utiliser les 2 premières composantes.

```python
x, y = 0,1
```

On peut maintenant créer le cercle des corrélations.

```python
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(0, pca_secto.components_.shape[1]):
    ax.arrow(0,
             0,  # Début de la flèche à l'origine
             pca_secto.components_[x, i],  #0 pour PC1
             pca_secto.components_[y, i],  #1 pour PC2
             head_width=0.05,
             head_length=0.05, 
             width=0.02,              )

    plt.text(pca_secto.components_[x, i] + 0.05,
             pca_secto.components_[y, i] + 0.05,
             X_secto.columns[i])
    
# affichage des lignes horizontales et verticales
plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')


# nom des axes, avec le pourcentage d'inertie expliqué
plt.xlabel('PC{} ({}%)'.format(x+1, round(100*pca_secto.explained_variance_ratio_[x],1)))
plt.ylabel('PC{} ({}%)'.format(y+1, round(100*pca_secto.explained_variance_ratio_[y],1)))

# titre
plt.title("Cercle des corrélations (PC{} et PC{})".format(x+1, y+1), fontsize=16, fontweight="bold")

#ajout du cercle
plt.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), np.sin(np.linspace(0, 2 * np.pi, 100)), linewidth=2)

plt.show()
```

Nous avons ici la première composante PC1 qui est une variable synthétique des variables nourriture, production et importations.
La deuxième composante est elle représentative de la variable "%Evolution de la nourriture".

### Analyse en Composantes Principales

```python
#On enregistre les clusters dans un df
X_cluster = X["Cluster"].to_frame()
#On supprime la colonne "Cluster" en prévision de l'ACP
X.drop(columns="Cluster",inplace=True)
```

On peut maintenant faire notre ACP.

```python
#Choix du nombre de composantes
n_components = 10

pca = PCA(n_components=n_components)
pca.fit(X)
```

#### Eboulis des valeurs propres

```python
pca.explained_variance_ratio_
```

On peut calculer notre variance cumulée et effectuer une visualisation:

```python
#Calcul de la variance cumulée
scree = (pca.explained_variance_ratio_*100).round(2)
scree_cum = scree.cumsum().round()
x_list = range(1, n_components+1)

#Représentation graphique
plt.figure(figsize=(10,3))
plt.bar(x_list, scree)
plt.plot(x_list, scree_cum,c="red",marker='o')
plt.xlabel("Rang de l'axe d'inertie")
plt.ylabel("Pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)
```

On voit ici que les deux premières composantes rassemblent un peu plus de 50% de la variance.

#### Cercle des corrélations

Intéressons nous dans un premier temps aux différentes composantes.

```python
#Calcul des coéfficients des variables pour chaque composante
pcs = pca.components_
#On met ces valeurs dans un df
pcs = pd.DataFrame(pcs)
#On rajoute les colonnes
pcs.columns = X.columns
pcs.index = [f"PC{i}" for i in x_list]
#On transpose pour une meilleure visibilité
pcs = pcs.T
pcs
```

```python
#Heatmap
sns.heatmap(pcs, cmap="YlGnBu")

plt.xlabel("Composantes")
plt.ylabel("Variables")

plt.show()
```

```python
#On enregistre les colonnes dans une variable
features_acp = pcs.columns
```

Définissons nos axes x et y. 
Nous allons utiliser les 2 premières composantes.

```python
x, y = 0,1
```

Nous pouvons maintenant créer le cercle des corrélations.

```python
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Début de la flèche à l'origine
             pca.components_[x, i],  #0 pour PC1
             pca.components_[y, i],  #1 pour PC2
             head_width=0.07,
             head_length=0.07, 
             width=0.02,              )

    plt.text(pca.components_[x, i] + 0.05,
             pca.components_[y, i] + 0.05,
             X.columns[i])
    
# affichage des lignes horizontales et verticales
plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')


# nom des axes, avec le pourcentage d'inertie expliqué
plt.xlabel('PC{} ({}%)'.format(x+1, round(100*pca.explained_variance_ratio_[x],1)))
plt.ylabel('PC{} ({}%)'.format(y+1, round(100*pca.explained_variance_ratio_[y],1)))

# titre
plt.title("Cercle des corrélations (PC{} et PC{})".format(x+1, y+1), fontsize=16, fontweight="bold")

#ajout du cercle
plt.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), np.sin(np.linspace(0, 2 * np.pi, 100)), linewidth=2)

plt.show()
```

Regardons maintenant les composantes 3 et 4:

```python
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Début de la flèche à l'origine
             pca.components_[2, i],  #0 pour PC1
             pca.components_[3, i],  #1 pour PC2
             head_width=0.07,
             head_length=0.07, 
             width=0.02,              )

    plt.text(pca.components_[2, i] + 0.05,
             pca.components_[3, i] + 0.05,
             X.columns[i])
    
# affichage des lignes horizontales et verticales
plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')


# nom des axes, avec le pourcentage d'inertie expliqué
plt.xlabel('PC{} ({}%)'.format(2+1, round(100*pca.explained_variance_ratio_[2],1)))
plt.ylabel('PC{} ({}%)'.format(3+1, round(100*pca.explained_variance_ratio_[3],1)))

# titre
plt.title("Cercle des corrélations (PC{} et PC{})".format(2+1, 3+1), fontsize=16, fontweight="bold")

#ajout du cercle
plt.plot(np.cos(np.linspace(0, 2 * np.pi, 100)), np.sin(np.linspace(0, 2 * np.pi, 100)), linewidth=2)

plt.show()
```

Ici l'interprétation des différentes composantes est rendu difficile par le fait que nous ayons des variables très différentes avec notamment des données quantitatives (production, importations, nourriture, PIB par habitant), des données mesurées en % (taux d'inflation, taux de chomage), un indice et des évolutions en % (population, consommation de volailles, PIB).

Nous pouvons toutefois identifier une certaine tendance selon les composantes.

La première composante (PC1) nous montre l'impact des variables suivantes:

- Indice de facilité des affaires
- Croissance du PIB
- PIB par habitant
- Taux d'inflation

PC1 est donc une **variable synthétique des indicateurs économiques.**

La seconde composante (PC2) montre l'impact des variables:

- Nourriture (consommation de viande de volailles)
- Production
- Importations (dans une moindre mesure)

PC2 est donc une **variable synthétique des données sectorielles.**

Les données relatives à l'évolution de la consommation de viande de volailles et au taux d'inflation sont clairement sur la composante PC3, tandis que les données relatives au taux de chômage sont sur la composante F4.
